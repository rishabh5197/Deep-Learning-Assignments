```python
import matplotlib.pyplot as plt
```

```python
def plotting(name,label):
    value=[]
    for i in range(-10,11):
        value.append(name(i))
    plt.figure(figsize=(16,8))
    plt.axhline(color="green")
    plt.axvline(color="orange")
    plt.plot(value,color="red",label=label)
    plt.grid()
    plt.legend()
    plt.title(label)
    plt.show()
```

#### Initializing tanh activation function

```python
def tanh(value): 
    return 2/(1+2.7182818284590452353602874713527**(-2*value))
```

```python
plotting(tanh,'tanh')
```

#### Initializing erelu activation function

```python
def erelu(value,alpha=1):
    if value>0:
        return value
    else:
        return alpha*(2.7182818284590452353602874713527**(value) - 1)
```

```python
plotting(erelu,'erelu')
```

#### Initializing sigmoid activation function

```python
def sigmoid(value):
    return 1/(1+2.7182818284590452353602874713527**-(value))
```

```python
plotting(sigmoid,'sigmoid')
```

#### Initializing lrelu activation function

```python
def lrelu(value):
    if value<0:
        return value*0.01
    else:
        return value
```

```python
plotting(lrelu,"lrelu")
```

#### Initializing relu activation function

```python
def relu(value):
    if value<0:
        return 0
    else:
        return value
```

```python
plotting(relu,"relu")
```
